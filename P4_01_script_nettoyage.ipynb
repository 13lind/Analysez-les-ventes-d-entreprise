```python
#Importations des librairies

import pandas as pd
import numpy as np
```

# dataframe products

```python
# importation csv file

products = pd.read_csv('products.csv')
```

```python
# visulation products
products.head(5)
```

```python
# chngement nom des colonnes

products.rename(columns={'price':'prix', 'categ' : 'categ_prod'},inplace=True)
products
```

```python
# Renomage category 0,1,2 avec list 

Category_produits = []

for i in np.arange(len(products.index)):
        if products.categ_prod[i] == 0:
            Category_produits.append("livre")
            
        elif products.categ_prod[i] == 1:
            Category_produits.append("decoration")
            
            
        else: 
                 Category_produits.append("loisir_creatifs")
        
# Transformer les listes en Dataframe Tmp, et ensuite lier avec dataframe d'origine

DF_TmpAnnee = pd.DataFrame({'category_produits' :  Category_produits})
products = pd.concat([products,DF_TmpAnnee], axis=1)
products
```

```python
# avec methode describe je trouve que il y a -1 valeur dans colonne Prix

products.describe()
```

```python
# j'ai netoyée le valeur -1

products = products.query("~(prix < 0)")
products.describe()
```

```python
#verification des doublons
products['id_prod'].duplicated().sum()
```

```python
#verification des doublons

products.duplicated().sum()
```

```python
# et je garde to _csv ces changements

products.to_csv("products.csv")
```

# dataframe customers

```python
# importation csv file

customers = pd.read_csv('customers.csv')
```

```python
# visulazion


customers.head(5)
```

```python
#  Renomage des colonnes

customers.rename(columns={'sex':'gendre', 'birth' : 'annee_de_naissance'},inplace=True)
```

```python
# renomage f = femme , m = musculin


customers['gendre'] = np.where((customers.gendre == 'f'), 'femme', customers.gendre)
customers['gendre'] = np.where((customers.gendre == 'm'), 'homme', customers.gendre)
```

```python
# verification des colonnes et des rows
customers.shape
```

```python
# verification doublon

customers.duplicated('client_id').sum()
```

```python
# on ajoute colonne Age - calcul annee precedente moins annee de naissance
customers['age'] = 2021 - customers['annee_de_naissance']
customers.sort_values(by = 'age', ascending=False)
```

```python
# ajoute une colonne avec tranche de age 

Tranche_de_Age = []

for i in np.arange(len(customers.index)):
        if customers.annee_de_naissance[i] <=1951>=1929:
            Tranche_de_Age.append("(70,70+]")
            
        elif customers.annee_de_naissance[i] <= 1971>=1952:
                Tranche_de_Age.append("(50,69]")
        elif customers.annee_de_naissance[i] <= 1991>=1972:
                Tranche_de_Age.append("(30,49]")
            
        else: 
                Tranche_de_Age.append("(17,29]")
        
# Transformer les listes en Dataframe Tmp, et ensuite lier avec dataframe d'origine

DF_TmpAnnee = pd.DataFrame({'tranche_de_age' : Tranche_de_Age})
customers = pd.concat([customers,DF_TmpAnnee], axis=1)
customers.head(15)
```

```python
# ajoute une colonne avec classement de age
Classement_age = []

for i in np.arange(len(customers.index)):
        if customers.annee_de_naissance[i] <=1951>=1929:
            Classement_age.append("4-Senior")
            
        elif customers.annee_de_naissance[i] <= 1971>=1952:
            Classement_age.append("3-Cinq_Senior")
        elif customers.annee_de_naissance[i] <= 1991>=1972:
            Classement_age.append("2-Adulte")
            
        else: 
                 Classement_age.append("1-Jeune Adulte")
        
# Transformer les listes en Dataframe Tmp, et ensuite lier avec dataframe d'origine

DF_TmpAnnee = pd.DataFrame({'classement_age' :  Classement_age})
customers = pd.concat([customers,DF_TmpAnnee], axis=1)
customers.head(15)
```

```python
# to_csv enregistre chaque changement

customers.to_csv("customers.csv")
```

```python
# Verification combien il y a Test transactions

transactions.isin(['T_0']).sum()
```

```python
#  change colonne dtype colonne date dans format datetime64
transactions['date'] = pd.to_datetime(transactions['date'])
transactions.info()
```

```python
#  change format dans format "%Y-%m-%d %H:%M:%S"
# je separe colonne date en Date et Time

transactions['Time'] = transactions["date"].dt.strftime("00:%M:%S.%f")

transactions['Date'] = transactions["date"].dt.strftime("%Y-%m-%d")
```
